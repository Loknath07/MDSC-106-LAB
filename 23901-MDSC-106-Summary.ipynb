# Stacked Bar Plots

```python
x = [ 'A', 'B', 'C', 'D']
y1 = [10,20,10,30]
y2 = [20,25,15,25]             # we created a data

plt.bar(x,y1, color='r')
plt.bar(x,y2, bottom=y1 , color='b')
plt.show()                    # plot bars in stack manner
```

```python
x = [ 'A', 'B', 'C', 'D']
y1 =np.array( [10,20,10,30])
y2 =np.array( [20,25,15,25] )
y3 =np.array([12,15,19,6])
y4 =np.array([10,29,13,19])  # data increased

plt.bar(x,y1, color='r')
plt.bar(x,y2, bottom=y1, color='b')
plt.bar(x,y3, bottom=y1+y2, color='g')
plt.bar(x,y4, bottom=y1+y2+y3, color='y')
plt.xlabel("Teams")
plt.ylabel("Score")
plt.legend(["Round 1", "Round 2", "Round 3", "Round 4"])
plt.title("Scores by team in four rounds")
plt.show()
```

###### By making patch_artist=True we can give a color to box plots

```python
np.random.seed(10)
data1=np.random.normal(100,10,200)
data2=np.random.normal(90,20,200)
data3=np.random.normal(80,30,200)
data4=np.random.normal(70,40,200)
data= [data1,data2,data3,data4]

fig=plt.figure(figsize=(10,7))

ax= fig.add_axes([0,0,1,1])
bp = ax.boxplot(data , patch_artist=True )
plt.show()
```

1) The middel line in box says the median.
2) If the tail is more (box down and up stright line) that is skewed.
3) If the tail is more to down then it is right skewed.
4) if the tail is more to up then it is left skewed.
