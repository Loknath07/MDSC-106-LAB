```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

D1 = pd.read_csv("TelcomCustomer-Churn_1.csv")
D1
```

```python
D2 = pd.read_csv("TelcomCustomer-Churn_2.csv")
D2
```

```python
merged_df = D1.merge(D2,on='customerID')
```

```python
merged_df
```

```python
columns = merged_df.columns.tolist()
print(columns)
```

```python
expected_columns =['customerID', 'gender', 'SeniorCitizen', 'Partner', 'Dependents', 'tenure', 'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod', 'MonthlyCharges', 'TotalCharges', 'Churn']
columns_incorporated = set(merged_df.columns)==set(expected_columns)
if columns_incorporated:
    print("All columns are incorporated in the merged DataFrame.")
else:
    print("Not all columns are incorporated in the merged DataFrame.")
```

```python
col = merged_df.select_dtypes(include=['float', 'int']).columns
for column_name in merged_df:
    if(column_name != 'customerID' and  column_name not in col):
        x = merged_df[column_name].value_counts()
        plt.pie(x, labels=x.index, autopct='%1.1f%%', startangle=140)
        plt.title(f'{column_name}')
        plt.show()
```
